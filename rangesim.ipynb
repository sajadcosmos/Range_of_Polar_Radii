```
import numpy as np
from random import uniform
range_values = list()
for j in range(1000):
  d_values = []
  for i in range(10):
    x = uniform(-1,1)
    y = uniform(-1,1)
    d = np.sqrt(x*x+y*y)
    if d<=1:
      d_values.append(d)
  range_value = max(d_values) - min(d_values)
  range_values.append(range_value)
print(np.average(range_values))
```

```
x_sim = range(3,100)
y_sim = []
for k in x_sim:
  range_values = []
  for j in range(1000):
      d_values = []
      for i in range(k):
          r = np.sqrt(uniform(0,1))
          theta = uniform(0, 2*np.pi)
          x = r*np.cos(theta)
          y = r * np.sin(theta)
          d = np.sqrt(x**2+y**2)
          d_values.append(d)
      range_value = max(d_values) - min(d_values)
      range_values.append(range_value)
  y_sim.append(np.average(range_values))
```

```
plt.xlabel("No. of points")
plt.ylabel("Value of range")
plt.plot(x_sim, y_sim)
plt.show()
```
